```python
# example 1.1 - a linked list class.
class LinkedListNode():
    '''node that maintains data and next_node references'''
    
    # __init__ defines the class's constructor.
    # default values can be provided for functions.
    def __init__(self, next_node = None, data = None):
        self.next_node = next_node
        self.data = data
        
        
# example 1.2 - using the linked list class to create and search through a list.
x3 = LinkedListNode(data=3)  # example of a keyword parameter.
x2 = LinkedListNode(x3,2)
x1 = LinkedListNode(x2,1) 

current_node = x1
target = 3
print("Searching for {}".format(target))

# this code that actually loops through the linked list
while current_node is not None and current_node.data != target:
    current_node = current_node.next_node
    
if current_node is not None:
    print("Did we find node x3? {}".format(current_node == x3))
else:
    print("The target was not found in the list")
print("")    
    

current_node = x1
target = 4
print("Searching for {}".format(target))
while current_node is not None and current_node.data != target:
    current_node = current_node.next_node
if current_node is not None:
    print("Did we find node x3? {}".format(current_node == x3))
else:
    print("The target was not found in the list")
```

```python
# example 3.1 - a stack class

class Stack():
    '''basic stack implementation'''
    
    def __init__(self):
        self.list = []
        
    def push(self, item):
        self.list.append(item)
        
    def pop(self):
        return self.list.pop(-1)
    
    # note: this function is used by len()
    def __len__(self):
        return len(self.list)

    
# example 3.2 - using the Stack class
x = Stack()
x.push(3)
x.push(2)
x.push(1)
print("length: {}".format(len(x)))
print("pops: {}, {}, {}".format(x.pop(), x.pop(), x.pop()))
```

```python
# example 4.1 - an example of a simple hash map class
class HashMap():
    '''basic hash map implementation'''
    
    def __init__(self, base_size = 127):
        # this is a list comprehension, check them out, they're cool.
        self.arr = [None for i in xrange(base_size)]
        self.base_size = base_size
    
    def add_item(self, key, data):
        # compute the index from a hash value of the key (maps to uniform dist.)
        index = abs(hash(key)) % self.base_size
        if self.arr[index] is not None:
            raise KeyError('Key already in use')
        self.arr[index] = data
        
    def get_item(self, key):
        index = abs(hash(key)) % self.base_size
        return self.arr[index]
    
    
# example 4.2 - using a hash map
x = HashMap()
x.add_item('yellow',121)
x.add_item('blue',333)
print('getting item "yellow" = {}'.format(x.get_item('yellow')))
print('getting item "blue" = {}'.format(x.get_item('blue')))
print('getting item "orange" = {}'.format(x.get_item('orange')))
print("")


# example 4.3 - the python dict class, {}, is a better implementation
x = {}
x['yellow'] = 121
x['blue'] = 333
print('getting item "yellow" = {}'.format(x['yellow']))
print('getting item "blue" = {}'.format(x['blue']))
# dicts throw an error for keys that are not present
```

```python
# example 5.1 - a binary search tree
class Node():
    '''a node in a binary tree'''
    def __init__(self,data=None,left=None,right=None):
        self.data = data
        self.left = left
        self.right = right
    
    def __str__(self):
        return str(self.data)
    
    def __repr__(self):
        return str(self)

# the key concept to understand is a Node, but these can be used to build
# useful things like a binary search tree. For interest, I've included 
# a few additional methods for this class.
class BST():
    '''a binary search tree'''
    def __init__(self):
        self.root = None
    
    def add(self,data):
        '''add a new node with the given data'''
        if self.root is None:
            self.root = Node(data)
            return self.root
        current_node = self.root
        while current_node is not None:
            if data < current_node.data:
                # if we've found where the new node belongs, put it in place
                if current_node.left is None:
                    current_node.left = Node(data)
                    return current_node.left
                else:
                    # otherwise, move to the next node
                    current_node = current_node.left
            else:
                if current_node.right is None:
                    current_node.right = Node(data)
                    return current_node.right
                else:
                    current_node = current_node.right
    
    def find(self, data, also_return_parent=False):
        '''find a node, and optionally return it's parent as well'''
        # returning a parent is useful for deleting a node
        if self.root is None:
            return None
        if self.root.data == data:
            if also_return_parent:
                return (self.root, None)
            else:
                return self.root
        current_node = self.root
        while current_node is not None:
            if data < current_node.data:
                if data == target_node.left.data:
                    if also_return_parent:
                        return (target_node.left, target_node)
                    else:
                        return current_node.left
                else:
                    current_node = current_node.left
            else:
                if data == current_node.right.data:
                    if also_return_parent:
                        return (current_node.left, current_node)
                    else:
                        return current_node.right
                else:
                    current_node = current_node.right
        if also_return_parent:
            return (None, None)
        else:
            return None
        
    def delete(self,data):
        '''remove the given piece of data and return the node'''
        if self.root is None:
            return None
        # if finding the parent and target node
        target_node, targets_parent = self.find(data,also_return_parent=True)
        
        # find left most node and parent of right child
        next_node, nexts_parent = target_node.right, target_node
        while next_node.left is not None:
            next_node, nexts_parent = next_node.left, next_node
        
        # now replace target_node with next_node
        nexts_parent.left = next_node.right
        
        if targets_parent is None:
            self.root = next_node
        else:
            if target_node == targets_parent.left:
                targets_parent.left = next_node
            else:
                targets_parent.right = next_node
            
        next_node.left = target_node.left
        next_node.right = target_node.right
            
    @staticmethod
    def inorderlist(current_node,node_list):
        '''returns all of the nodes in sorted order'''
        if current_node is not None:
            BST.inorderlist(current_node.left, node_list)
            node_list.append(str(current_node.data))
            BST.inorderlist(current_node.right, node_list)
                
    # provides the method used to print out the tree as a string.
    def __str__(self):
        node_list = []
        BST.inorderlist(self.root, node_list)
        return ",".join(node_list)
    
    
# example 5.2 - using a binary search tree
x = BST()
x.add(5)
x.add(4)
x.add(3)
x.add(1)
print(x)
```

```python
# example 6.1 - a graph node class
class GraphNode():
    '''implementation of a graph node with a generator expression'''
    
    def __init__(self, value, neighbors = None):
        self.value = value
        if neighbors is None:
            self.neighbors = []
        else:
            self.neighbors = neighbors
        
    def add_neighbor(self, node):
        self.neighbors.append(node)
        
    def neighbors_group(self):
        # this is a generator expression that lets you iterate easily
        for neighbor in self.neighbors:
            yield neighbor
            
            
# example 6.2 - using the graph node class
x1 = GraphNode(1)
x2 = GraphNode(2)
x3 = GraphNode(3,[x1,x2])

# now use the generator to iterate over all the neighbors
for x in x3.neighbors_group():
    print(x.value)
```

```python
# example 7.1 - a priority queue class
class PriorityQueue():
    '''priority queue implementation that always returns the maximum'''
    
    def __init__(self):
        self.list = []
        
    def add(self,value):
        new_index = len(self.list)
        self.list.append(value)
        next_index = new_index / 2  # automatically rounds down
        while self.list[next_index] < self.list[new_index]:
            # swap the two positions
            self.list[next_index],self.list[new_index] = (
                self.list[new_index], self.list[next_index])
            new_index = next_index
            next_index = new_index / 2
            # automatically stops at zero
    
    def pop(self):
        r_val = self.list[0]
        if len(self.list) == 1:
            self.list.pop(0)
        else:
            # move the last position to the front and restore order.
            # compare newly moved value to the ones that are supposed to be 
            # beneath and move as appropriate.
            self.list[0] = self.list.pop(-1)
            next_pos = 1
            while (next_pos + 1 < len(self.list) and
                    max(self.list[next_pos], self.list[next_pos + 1]) >
                    self.list[next_pos / 2]):
                # always move the highest value forward.
                if self.list[next_pos] > self.list[next_pos + 1]:
                    self.list[next_pos / 2], self.list[next_pos] = (
                        self.list[next_pos], self.list[next_pos / 2])
                    next_pos = next_pos * 2
                else:
                    self.list[next_pos / 2], self.list[next_pos + 1] = (
                        self.list[next_pos + 1], self.list[next_pos / 2])
                    next_pos = (next_pos + 1) * 2
            # check to make sure an uneven tail is still corrected for.
            if (next_pos < len(self.list) and 
                    self.list[next_pos] > self.list[next_pos / 2]):
                self.list[next_pos / 2], self.list[next_pos] = (
                        self.list[next_pos], self.list[next_pos / 2])
        return r_val

    def __len__(self):
        return len(self.list)
    
# example 7.2 - using the priority queue
x = PriorityQueue()
for y in [1,2,5,3,4,6]:
    x.add(y)
print(x.list)
while len(x) > 0:
    print(x.pop())
```

```python
# example 8.1 - binary search
ordered_list = range(100)  # all numbers 1-100.

top_ind = 99
bottom_ind = 0
target_value = 37

print('the loop method')
current_ind = (top_ind + bottom_ind) / 2
while ordered_list[current_ind] != target_value:
    print(current_ind)
    if ordered_list[current_ind] < target_value:
        bottom_ind = current_ind
    else:
        top_ind = current_ind
    current_ind = (top_ind + bottom_ind) / 2
    
print("index {}: value {}".format(current_ind, ordered_list[current_ind]))
print('')

print('the recursive method')
def binary_search(list, top, bottom, target):
    current_ind = (top + bottom) / 2
    print(current_ind)
    if ordered_list[current_ind] == target:
        return current_ind
    elif ordered_list[current_ind] < target:
        bottom = current_ind
    else:
        top = current_ind
    return binary_search(list, top, bottom, target)
    
top_ind = 99
bottom_ind = 0
target_value = 37
result = binary_search(ordered_list, top_ind, bottom_ind, target_value)
print("index {}: value {}".format(result, ordered_list[result]))
```

```python
# example 9.1 - tree for BFS and DFS

class Node():
    def __init__(self, label):
        self.label = label
        self.children = []
        
node_list = []
for x in xrange(7):
    node_list.append(Node(x))
node_list[0].children = [node_list[1], node_list[2]]
node_list[1].children = [node_list[3], node_list[4]]
node_list[2].children = [node_list[5], node_list[6]]

# example 9.2 - DFS - recursive
def print_dfs(node):
    print(node.label)
    for child in node.children:
        print_dfs(child)
print('DFS - recursive')        
print_dfs(node_list[0])
print('')

# example 9.3 - DFS - stack method
print('DFS - stack')
nodes_to_visit = [node_list[0]]
while nodes_to_visit:
    current_node = nodes_to_visit.pop(-1)
    print(current_node.label)
    nodes_to_visit.extend(current_node.children)
print('')

# example 9.3 - BFS - queue method
print('BFS - queue')
nodes_to_visit = [node_list[0]]
while nodes_to_visit:
    current_node = nodes_to_visit.pop(0)
    print(current_node.label)
    nodes_to_visit.extend(current_node.children)
print('')
```

=====================================

```python
# example 11.1 - quicksort algorithm

def quicksort(list, top, bottom):
    if top - bottom < 1:
        return
    pivot = list[top]  # choose an arbitrary pivot, that must be moved to the end
    current_swap_index = bottom
    for index in xrange(bottom,top):
        if list[index] < pivot:
            list[index], list[current_swap_index] = (
                list[current_swap_index], list[index])
            current_swap_index += 1
    list[top], list[current_swap_index] = list[current_swap_index], list[top]
    print(list)
    quicksort(list, top, current_swap_index + 1)
    quicksort(list, current_swap_index - 1, bottom)


# example 11.2 - how quicksort runs on different lists

unsorted_list = [6, 3, 0, 2, 4, 5, 1]

new_list = unsorted_list[:]

quicksort(new_list, len(new_list) - 1, 0)
print('')

print('with a reversed list')
reversed_list = [6, 5, 4, 3, 2, 1, 0]
new_list = reversed_list[:]
quicksort(new_list, len(new_list) - 1, 0)
print('')

print('with a nearly sorted list')
nearly_sorted_list = [1,2,3,4,5,6, 0]
new_list = nearly_sorted_list[:]
quicksort(new_list, len(new_list) - 1, 0)
print('')
```

<a id="merge_sort"></a> merge sort
=====================================

```python
# example 12.1 - merge sort algorithm

def merge_sort(list):
    if len(list) < 2:
        # print(list)
        return list
    left = merge_sort(list[:len(list)/2])
    right = merge_sort(list[len(list)/2:])
    new_list = []
    while left and right:
        if left[0] < right[0]:
            new_list.append(left.pop(0))
        else:
            new_list.append(right.pop(0))
    new_list.extend(left)
    new_list.extend(right)
    print(new_list)
    return new_list

# example 12.2 - how merge sort runs on different lists

unsorted_list = [6, 3, 0, 2, 4, 5, 1]

new_list = unsorted_list[:]

merge_sort(new_list)
print('')

print('with a reversed list')
reversed_list = [6, 5, 4, 3, 2, 1, 0]
new_list = reversed_list[:]
merge_sort(new_list)
print('')

print('with a nearly sorted list')
nearly_sorted_list = [1,2,3,4,5,6, 0]
new_list = nearly_sorted_list[:]
merge_sort(new_list)
print('')
```

```python
# example 13.1 - a memoization function
def memoize(obj):
    '''a decorator to memoize a function's positional arguments'''
    cache = {}
    obj.cache = cache
    def memoizer(*args):
        if args not in cache:
            cache[args] = obj(*args)
        return cache[args]
    return memoizer

# example 13.2 - the effect of memoization on an expensive function
import time
sleep_time = 0.1
calculate_number = 7

def expensive_fibonacci(x):
    time.sleep(sleep_time)  # pretend this is very hard to compute
    if x <= 1:
        return x
    return expensive_fibonacci(x-1)+expensive_fibonacci(x-2)

start_time = time.time()
value = expensive_fibonacci(calculate_number)
duration = time.time() - start_time
print("expensive_fibonacci returned {} after {} seconds".format(value, duration))

start_time = time.time()
value = expensive_fibonacci(calculate_number)
duration = time.time() - start_time
print("expensive_fibonacci returned {} after {} seconds".format(value, duration))

@memoize
def memoized_fibonacci(x):
    time.sleep(sleep_time)  # pretend this is very hard to compute
    if x <= 1:
        return x
    return memoized_fibonacci(x-1)+memoized_fibonacci(x-2)

start_time = time.time()
value = memoized_fibonacci(calculate_number)
duration = time.time() - start_time
print("memoized_fibonacci returned {} after {} seconds".format(value, duration))


start_time = time.time()
value = memoized_fibonacci(calculate_number)
duration = time.time() - start_time
print("memoized_fibonacci returned {} after {} seconds".format(value, duration))
```

```python
# example 14.1 - putting a set of items in to one or more knapsacks

# knapsack problem - put as much weight into sacks as possible
sacks = [18,4,2,2,4]
items = [17,3,1,2,2,4]
contents = [[] for sack in sacks]

def fill_sacks(sacks, items, contents, total = 0):
    #print( sacks, items, contents, total)
    # base cases - no more items, no space in sacks, return our accomplishment
    if are_no_actions(sacks, items):
        return total, contents
    # recursive cases - try to add each item to the first sack and recurse
    # return the best set of contents
    subproblems = []
    for i, sack in enumerate(sacks):
        for j, item in enumerate(items):
            subresult = try_to_add_item(sacks, items, contents, i, j, total)
            if subresult is not None:
                subproblems.append(subresult)
    return max(subproblems, key=lambda x:x[0])
    
def are_no_actions(sacks, items):
    '''returns true if no possible progress can be made'''
    if not items or not sacks:
        return True
    smallest_item = min(items)
    largest_sack = max(sacks)
    return smallest_item > largest_sack

def try_to_add_item(sacks, items, contents, sack_index, item_index, total):
    if items[item_index] > sacks[sack_index]:
        return None
    new_sacks = list(sacks)
    new_sacks[sack_index] -= items[item_index]
    new_items = [item for j, item in enumerate(items) if j != item_index]
    new_contents = [list(sack_contents) for sack_contents in contents]
    new_contents[sack_index].append(items[item_index])
    new_total = total + items[item_index]
    return fill_sacks(new_sacks, new_items, new_contents, new_total)
    
fill_sacks(sacks, items, contents)


# example 14.2 - endless items in to one or more knapsacks

from pprint import pprint

# knapsack problem 2 - put as much value into a sack as possible given the weights
max_weight = 24
items = [
   (32, 10000),
   (10, 20),
   (4, 4),
   (1, 1.9),
]

def fill_knapsack(max_weight, items):
    # Try putting an item into the sack, report change in value
    @memoize
    def best(index,cur_weight):
        #print(index,cur_weight)
        # base case
        if index >= len(items):
            return 0
        weight, value = items[index]
        if weight > cur_weight:
            return best(index + 1, cur_weight)
        # recurse to determine if it is better to add or skip
        return max(best(index, cur_weight - weight) + value, # limit to 0/1 by using i+1
                   best(index + 1, cur_weight))
        
    # Initialize variables.
    contents = []
    current_weight = max_weight
    # Step through and test whether adding an item is better than skipping that item.
    # To be efficient, best must be memoized (or put in matrix form). The first call
    # fills out the matrix, but then we want to read it off by testing every step.
    current_index = 0
    depth = 0
    while current_index < len(items):
        weight, value = items[current_index]
        #print(current_index,current_weight, depth)
        while(weight <= current_weight and
              (best(current_index, current_weight - weight) + value)
              > (best(current_index + 1, current_weight))):
            #print('adding ' + str(items[current_index]))
            contents.append(items[current_index])
            current_weight -= weight
        current_index += 1
    #pprint(best.cache)
    return contents

results = (fill_knapsack(max_weight, items))
print(results)
print(sum([x[0] for x in results]))
print(sum([x[1] for x in results]))
```
